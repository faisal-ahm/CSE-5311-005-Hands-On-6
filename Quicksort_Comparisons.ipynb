```python
import time
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def partition(arr, low, high):
    pivot = arr[high]  # Choose the last element as pivot
    i = low - 1  

    for j in range(low, high):
        if arr[j] <= pivot: 
            i += 1
            arr[i], arr[j] = arr[j], arr[i]  

    arr[i + 1], arr[high] = arr[high], arr[i + 1]  
    return i + 1


def quicksort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quicksort(arr, low, pi - 1)
        quicksort(arr, pi + 1, high)

# Example:
arr = [10, 7, 8, 9, 1, 5]
quicksort(arr, 0, len(arr) - 1)
print("Sorted array:", arr)
```

```python
import random

def random_partition(arr, low, high):
    rand_index = random.randint(low, high)  # Random pivot selection
    arr[high], arr[rand_index] = arr[rand_index], arr[high]
    return partition(arr, low, high)


def random_quicksort(arr, low, high):
    if low < high:
        pi = random_partition(arr, low, high)
        random_quicksort(arr, low, pi - 1)
        random_quicksort(arr, pi + 1, high)

# Example usage:
arr = [10, 7, 8, 9, 1, 5]
random_quicksort(arr, 0, len(arr) - 1)
print("Sorted array (random pivot):", arr)
```

```python
def benchmark_quicksort(array_generator, sizes):
    timings = []
    for n in sizes:
        print("size: ", n)
        arr = array_generator(n)
        start_time = time.time()
        random_quicksort(arr, 0, len(arr) - 1)
        end_time = time.time()
        timings.append((end_time - start_time) * 1e6)  # Convert to microseconds
    return timings

# Input generators
def best_case_input(n):
    return list(range(n))  # Sorted array

def worst_case_input(n):
    return list(range(n, 0, -1))  # Reverse sorted array

def average_case_input(n):
    return random.sample(range(n), n)  # Random array

# Smoothing function: Moving Average
def smooth_data(data, window_size=5):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Array sizes
sizes = range(1, 15001)  

# Benchmarking
best_case_timings = benchmark_quicksort(best_case_input, sizes)
worst_case_timings = benchmark_quicksort(worst_case_input, sizes)
average_case_timings = benchmark_quicksort(average_case_input, sizes)

# Smoothing the data with a moving average
window_size = 5  
smoothed_best_case = smooth_data(best_case_timings, window_size)
smoothed_worst_case = smooth_data(worst_case_timings, window_size)
smoothed_average_case = smooth_data(average_case_timings, window_size)


smoothed_sizes = list(sizes)[window_size - 1:]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(smoothed_sizes, smoothed_best_case, label='Best Case', color='green')
plt.plot(smoothed_sizes, smoothed_worst_case, label='Worst Case', color='red')
plt.plot(smoothed_sizes, smoothed_average_case, label='Average Case', color='blue')
plt.xlabel('Array Size (n)')
plt.ylabel('Time (microseconds)')
plt.title('Smoothed Quicksort Benchmarks: Best, Worst, and Average Case')
plt.legend()
plt.grid(True)
plt.show()

# NOTE: SCROLL TO THE BOTTOM OF THE OUTPUT TO SEE THE UNSMOOTHED GRAPH
```
